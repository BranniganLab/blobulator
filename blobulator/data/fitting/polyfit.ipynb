```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
raw_data = pd.read_csv("all_humsaver_enrich_txt_h.csv", names=["L","H","dSNPs", "nSNPs", "SNPs", "Enrich", "p"], skiprows=1)
```

```python

mask = raw_data.H<=1

Ys = raw_data.loc[mask,"dSNPs"]
Ys=Ys/np.sum(Ys)

Xs = raw_data.loc[mask, ["H", "L"]]
Hs = Xs['H'].unique()
Ls = Xs['L'].unique()
```

```python
%matplotlib
histgrid = [[raw_data.loc[np.logical_and(raw_data.H==H,raw_data.L==L), 'dSNPs'] for L in Ls] for H in Hs]

plt.imshow(histgrid, origin='lower')
```

```python
%matplotlib
histgrid = [[raw_data.loc[np.logical_and(raw_data.H==H,raw_data.L==L), 'dSNPs'] for L in Ls] for H in Hs]
histgrid = np.cumsum(histgrid, axis=1)
histgrid = np.cumsum(histgrid, axis=0)

plt.imshow(histgrid, origin='lower')
```

```python
def get_dsnp(data, Hs, Ls):
    vals = []
    for H in Hs:
        toadd = []
        for L in Ls:
            mask = np.logical_and(data.H==H, data.L==L)
            toadd.append(data.dSNPs.loc[mask])
        vals.append(toadd)
    return vals
```

```python
# Make data.
Z = np.array(get_dsnp(raw_data, Hs, Ls))
X, Y = np.meshgrid(Hs, Ls)
Z = Z/np.sum(Z)
```

```python
%matplotlib
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})


# Plot the surface.
surf = ax.scatter3D(X, Y, Z.T, c=Z.T,cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Customize the z axis.
#ax.set_zlim(0, 1)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
